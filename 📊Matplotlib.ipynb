# Introduction of Matplotlib

## Why Mathplotlib?

• Built on NumPy arrays (and Python)

• Integrates directly with pandas

• Can create basic or advanced plots

• Simple to use interface (once you get the foundations)

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

```python
# 1 Method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2 Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1,])
ax.plot(x , y)
plt.show()
```

```python
# 3 Method
fig, ax = plt.subplots()
ax.plot(x , y); 
```

```python
# Example

%matplotlib inline
import matplotlib.pyplot as plt

# Data
x = [1,2,3,6]
y = [10,20,30,40]

# setting plot
fig, ax = plt.subplots(figsize=(5,10)) #(width, hieght)

# plot data
ax.plot(x,y)

# custom plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# Save
fig.savefig("sample.png")
```

## Making figures with numpy arrays.

1. Line 
2. Scatter 
3. Bar
4. Histogram
5. Subplots

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x , x **2);
```

```python
# Scatter plot

fig, ax = plt.subplots()
ax.scatter(x , np.exp(x))
```

```python
# scatter plot
fig, ax = plt.subplots()
ax.scatter(x , np.sin(x));
```

```python
Prices = {"apple": 10,
         "orange": 5,
         "banana": 2}
fig, ax = plt.subplots()
ax.bar(Prices.keys(), Prices.values())
```

```python
fig, ax = plt.subplots()
ax.barh(list(Prices.keys()), Prices.values())
```

```python
x = np.random.randn(2000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
## Subploting options

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))
ax1.plot(x, x/2);
ax2.scatter(x = np.random.randn(100), y = np.random.randn(100));
ax3.barh(list(Prices.keys()), Prices.values())
ax4.hist(x);
```

```python
e = np.random.rand(5,4)
e
```

```python
fig, ax = plt.subplots()
ax.plot(e)
ax.scatter(e/2, e)
```

```python
fig, ax = plt.subplots(nrows=2,
               ncols=2,
               figsize= (10,5))

ax[0,0].plot(x , x/2)
ax[0,1].scatter(e, e/2.5)
ax[1,1].barh(list(Prices.keys()), Prices.values())
```
